```python
import pandas as pd
import numpy as np
petaks = pd.read_csv('data/pulpwood/plot_analysis.csv')
petaks2 = pd.read_csv('data/pulpwood/plot_analysis_new.csv')
print(petaks.shape[0])
print(petaks2.shape[0])
frames = [petaks, petaks2]
petaks=pd.concat(frames, ignore_index=True)
print(petaks.shape[0])
petaks
petaks.to_csv('data/pulpwood/plot_analysis_combined.csv')
```

```python
petaks = pd.read_csv('data/pulpwood/plot_analysis_combined.csv')
petaks['date'] = pd.to_datetime(petaks['date'])
petaks=petaks.loc[(petaks['age_at_obs_months']>0) & (petaks['age_at_obs_months']<49)]
petaks.columns
```

```python
max_indices = []
unique_dates = petaks['date'].unique()
unique_dates.sort()

veg_indices = {'ndvi', 'ndvi_80', 'ndvi_95', 'gndvi', 'gndvi_80', 'gndvi_95', 'evi','evi_80','evi_95', 'rvi','rvi_80', 'rvi_95','ndwi', 'ndwi_95', 'ndwi_80','ndwi_20','ndwi_05'}

for date in unique_dates:
    df = petaks.loc[(petaks['date']==date)]
    max_age = df['age_at_obs_months'].max()
    count = df['ndvi_count'].sum()
    obs_no = df.shape[0]    
    d = {'date':date, 'count': count, 'max_age' : max_age, 'obs_no': obs_no}
    
    for veg_index in veg_indices:
        max_index = df[veg_index].max()       
        d.update({veg_index: max_index})
    
    max_indices.append(d)

max_indices = pd.DataFrame.from_dict(max_indices)

for veg_index in veg_indices:
    df=max_indices.loc[max_indices['obs_no']>9, [veg_index]]
    max_indices[f"{veg_index}_adj"] = df.quantile(0.9)/df
max_indices = max_indices.fillna(1)
max_indices = max_indices.set_index('date')
max_indices
```

```python
age_dict = {"age1":[1,2],
            "age2":[3,4],
            "age3":[5,6],
            "age4":[7,9],
            "age5":[10,12],
            "age6":[13,24],
            "age7":[25,48],
           }
```

```python
quantiles = [0.01, 0.1, 0.25, 0.5, 0.75, 0.9, 0.99]
veg_indices = {'ndvi', 'ndvi_80', 'ndvi_95', 'gndvi', 'gndvi_80', 'gndvi_95', 'evi','evi_80','evi_95', 'rvi','rvi_80', 'rvi_95','ndwi', 'ndwi_95', 'ndwi_80','ndwi_20','ndwi_05'}

quantile_df = pd.DataFrame(index=quantiles)


for variant in {'Epel', 'Acra'}:
    for veg_index in veg_indices:
        for age in age_dict:
            base_index = veg_index.split('_')
            df = petaks.loc[(petaks['age_at_obs_months']>=age_dict[age][0])
                            & (petaks['age_at_obs_months']<=age_dict[age][1])
                            & (petaks['variant']==variant)
                            & (petaks[f'{base_index[0]}_count']>2000), [veg_index]]
            
            quantile_df[f"{veg_index}_{variant}_{age_dict[age][0]}_{age_dict[age][1]}"] = df.quantile(quantiles)
                                                             
quantile_df
```

```python
def get_threshold(veg_index, variant, quantile_df, threshold, ages):
    threshold_values = []
    for age in ages:
        if age >= 1 and age <=2:
            threshold_values = np.append(threshold_values, quantile_df.at[threshold, f"{veg_index}_{variant}_1_2"])
        elif age>= 3 and age <=4:
            threshold_values = np.append(threshold_values, quantile_df.at[threshold, f"{veg_index}_{variant}_3_4"])
        elif age >= 5 and age <=6:
            threshold_values = np.append(threshold_values, quantile_df.at[threshold, f"{veg_index}_{variant}_5_6"])
        elif age >= 7 and age <=9:
            threshold_values = np.append(threshold_values, quantile_df.at[threshold, f"{veg_index}_{variant}_7_9"])
        elif age >= 10 and age <=12:
            threshold_values = np.append(threshold_values, quantile_df.at[threshold, f"{veg_index}_{variant}_10_12"])
        elif age >= 13 and age <=24:
            threshold_values = np.append(threshold_values, quantile_df.at[threshold, f"{veg_index}_{variant}_13_24"])
        elif age >= 25 and age <=48:
            threshold_values = np.append(threshold_values, quantile_df.at[threshold, f"{veg_index}_{variant}_25_48"])
        else:
            threshold_values = np.append(threshold_values, np.nan)
    return threshold_values
```

```python
adjust = False

veg_indices = ['ndvi', 'ndvi_80', 'ndvi_95', 'evi', 'evi_95', 'gndvi', 'gndvi_95', 'rvi', 'rvi_95']

for plot_id in range(336,408):
    if not petaks.loc[petaks['plot_id']==plot_id].empty:
        subplot_number = 0
        plt.figure(plot_id, figsize=(30,5))
        
        for veg_index in veg_indices:
            

            df = petaks.loc[petaks['plot_id']==plot_id, ['date', 'planted_date', 'variant', 'age_at_obs_months', veg_index]]            
            
            if adjust:
                for i, row in df.iterrows():
                    df.at[i,veg_index] *= max_indices.at[row['date'], f"{veg_index}_adj"]            
                df.dropna()
            
            df = df.set_index('date')
            
            #df['rolling_max'] = df[veg_index].rolling('60d', min_periods=1).mean()
            df['rolling_max'] = df[veg_index].rolling('60d', min_periods=1).max()

            variant = df['variant'].iloc[0]
            planed_date = df['planted_date'].iloc[0]
            min_age = df.age_at_obs_months.min()
            max_age = df.age_at_obs_months.max()
                  
            subplot_number+=1
            plt.subplot(1,len(veg_indices),subplot_number)
            plt.plot(df.index.values, df[veg_index], 'kx')
            plt.plot(df.index.values, df['rolling_max'], 'b-')
                        
            #plot growing stage threshold
            threshold_low = get_threshold(veg_index=veg_index, variant=variant, threshold = 0.25, quantile_df=quantile_df, ages=df['age_at_obs_months'])
            threshold_high = get_threshold(veg_index=veg_index, variant=variant, threshold = 0.90, quantile_df=quantile_df, ages=df['age_at_obs_months'])
            df['threshold_low'] = threshold_low
            df['threshold_high'] = threshold_high

            df_temp = df.loc[(df['age_at_obs_months']<=12), ['threshold_low', 'threshold_high']]
            plt.plot(df_temp.index.values, df_temp['threshold_low'], color="green", linestyle='--')
            plt.plot(df_temp.index.values, df_temp['threshold_high'], color="red", linestyle='--')
            
            #plot mature stage threshold
            if adjust:
                threshold_low = get_threshold(veg_index=veg_index, variant=variant, threshold = 0.50, quantile_df=quantile_df, ages=df['age_at_obs_months'])
            else:
                threshold_low = get_threshold(veg_index=veg_index, variant=variant, threshold = 0.25, quantile_df=quantile_df, ages=df['age_at_obs_months'])
            threshold_high = get_threshold(veg_index=veg_index, variant=variant, threshold = 0.90, quantile_df=quantile_df, ages=df['age_at_obs_months'])
            df['threshold_low'] = threshold_low
            df['threshold_high'] = threshold_high
            df_temp = df.loc[(df['age_at_obs_months']>12), ['threshold_low', 'threshold_high']]
            plt.plot(df_temp.index.values, df_temp['threshold_low'], color="red", linestyle=':')
            plt.plot(df_temp.index.values, df_temp['threshold_high'], color="green", linestyle=':')
                        
            plt.xticks(rotation='vertical')
            plt.title(f"{veg_index}", fontsize=12)
            plt.title(f"{min_age}m", loc="left", fontsize=10)
            plt.title(f"{max_age}m", loc="right", fontsize=10)
            
        plt.suptitle(f"Plot: {plot_id}   Variant: {variant}  Adjusted: {adjust}")
        #plt.show()
        plt.savefig(f"data/pulpwood/plots/vi_{plot_id}_adjust_{adjust}.png")
        plt.close()
        
```
